```python
#문서를 벡터 인자로 불러오기

doc1 = "   고용관계론   사무직, 관리직, 비정규직, 첨단인력과 비노조 경영을 포괄하는 종업원과 경영자와의 사이에 근로 조건을 중심으로 서로 대립하는 관계를 합리적으로 해결하여 생산적 고용관계 정립을 위한 이론적 근거를 숙지하고 실무적 해결능력을 배양한다.   본 과목에서는 고용조건의 결정 및 준수를 위하여 노동자, 사용자, 정부가 형성하는 공식적 및 비공식적 관계를 의미하는 고용관계에 대한 이해를 증진시키고자 한다. 본 과목의 내용은 고용관계를 구성하는 기본요인, 고용관계의 주요이론과 한국 및 각국의 고용관계, 노동조합과 사용자 조직, 단체교섭과 단체협약, 쟁의조정과 부당노동행위, 비정규직의 고용관계 등으로 이루어져있다. 이를 통해 경영자로서 알아야 할 고용관계의 기본적인 지식과 실무능력을 배양함과 아울러 공인노무사와 같은 전문적인 자격증을 취득하기 위한 기본적인 과목이다
   경영정보시스템   조직운영 및 경영활동을 지원하는 경영정보시스템에 관련된 학문적 이론을 배우고, 정보시스템을 현장에서 이용하는데 발생하는 문제점들의 해결방안을 찾는다.   조직의 경영활동지원과 생산성 향상을 위해 이용되는 경영정보시스템(Management InformationSystems)의 이론과 실제적인 문제점들을 강의한다. 또한 기업의 경쟁우위를 창출하기 위한 정보통신기술의 전략적 활용방안과 정보시스템 및 정보자원 관리에 관하여 종합적으로 검토 연구한다.
   조직이론   조직행동기초이론을 배우고, 이를 활용하는 방법을 연구한다.   경영학의 기초가 되는 것은 통계학을 중심으로 한 수리적인 접근이 있고 다른 하나는 조직행동을 중심으로 한 개인과 집단의 움직임에 대한 역학을 파악하는 것으로 심리학, 사회학 문화인류학을 기업경영에 필요하도록 실용적으로 재편한 내용이다. Taylorism과 인간관계론을 재조명하고 Real Organization의 파악을 시도한다. 개인의 정체성 과 행동의 동기에 대한 이론과 현실에서의 의미를 학습한다
   소비자행동분석   소비자 행동의 다양한 측면을 이해토록 함으로써, 소비자를 대상으로 한 마케팅 활동에 시스템적 접근을 강구한다. 소비자의 정보처리 과정, 태도 형성, 구매의사결정과 관련된 주요 이론들을 이해한다.   소비자 행동분야에서 지금까지 이루어진 연구들과 기본사항들을 익히고 소비자 행동의 다양한 측면을 이해하도록 함으로써, 시장을 대상으로 한 마케팅 활동에 시스템적 접근을 강구한다.
   생산/물류관리   품질관리, 재고관리, 물류 분야에서 기업의 생산성을 개선하는 방안을 연구하고 이를 응용하는 방안을 배운다.   생산시스템의 설계, 계획 및 통제를 효과적으로 수행함으로써 기업의 생산성 향상 및 원가절감을 위한 최적방안을 검토한다. 품질관리, 재고관리, 물류에 중점을 두어 강의한다.
   중급회계Ⅱ   기업에 대한 정보를 외부의 이해관계자들에게 생산, 전달하는 재무회계의 이론적 근거를 숙지하고 부채, 자본, 수익과 비용 관련 회계실무처리 능력을 함양한다.   본 교과목은 회계정보를 산출하는 재무회계의 이론적 근거를 숙지하고 실무처리 능력을 높이도록 회계기준에 의한 회계처리 지침의 내용과 이론적 근거 및 회계처리 실무 내용을 학습한다. 본 과목을 수강하기 위해서는 회계원리 및 중급회계Ⅰ에 대한 선행학습이 요구되며, 교과 내용은 중급회계의 후반부인부채, 자본 및 수익과 비용회계의 이론적 회계처리 등으로 구성되어 있다. 본 과목은 경영자로서 기본적으로 알아야 할 재무회계에 대한 개념 및 기본지 식을 전달할 뿐만 아니라 공인회계사(CPA), 세무사(CTA), 재무분석사(CFA) 등 회계 및 재무 분야의 전문가 시험을 준비하는데 필요한 과목이다.
   세무회계   본 과목은 세무회계의 이론과 실무를 이해시키고 기업회계와 세법의 차이점을 비교 연구하여 기업이부담하는 법인세, 개인이 부담하는 소득세 및 상거래에서 발생하는 부가가치세법 등의 구체적인 계산과정을 학습한다.   본 과목은 세법규정과 기업회계기준의 차이 및 그 차이를 조정하는 세무조정에 중점을 둔다. 주요 내용에는 법인세 과세표준을 계산하기 위한 세무조정문제, 소득세회계, 부가가치세문제, 상속세 및 증여세의 문제가 포함된다. 따라서 본 과목은 세법개론의 학습을 바탕으로 법인세, 소득세, 부가가치세, 상속세 및 증여세를 산출할 수 있는 능력을 배양하는데 최종 목적이 있다. 본 과목은 기업의 세무 실무자뿐 아니라, 공인회계사(CPA), 세무사(CTA)등 세무 회계전문가의 시험을 준비하는 수험생에게는 반드시 이수해야 하는 필수적인 과목이다.
   통계학   통계학의 기초를 확립하고, 경영학의 타 분야에서 응용할 수 있는 통계기법을 연구한다.   확률론, 표본분포론, 추정 및 검정이론, 회귀분석 및 상관분석 등을 주 강의내용으로 한다. 다른 경영 분야에 응용될 수 있도록 경영문제의 통계적 분석기법을 함양시킨다.
   세법개론   본 과목은 기업의 경영과 개인의 일상생활에서 중요성이 증대되고 있는 조세에 대한 기초적인 개념과 체계 및 조세의 부과와 징수 방법 등을 학습한다.   본 과목에서는 조세의 전반적인 개요 및 세법의 기본원리와 해석 및 그 적용방법 등을 다루고 있는 국세기본법, 법인이 부담하는 조세를 다룬 법인세법, 상거래에서 발생하는 부가가치세법과 개인(사업자)이 부담하는 조세를 다룬 소득세법 등을 공부한다. 본 과목을 통해 조세의무 전반에 대한 기본소양을 습득할 수 있으며, 기업의 조세전략을 수립하는 데 필요한 기초지식을 습득할 수 있을 것이다. 본 과목은 기업의 경영자와 개인에게도 중요하며, 특히 공인회계사(CPA), 세무사(CTA)등 세무회계전문가가 되기 위해서는 반드시 이수해야 하는 기본적인 과목이다.
   경영혁신과 정보기술   경영혁신기법과 관련된 경영학이론을 연구하고 경영혁신기법의 실행을 위해 요구되는 정보통신 기술의 역할과 적용 방법을 학습한다.   경영학에서 새로이 대두된 최신 경영혁신 이론들을 분석하고, 기업이 급변하는 경영환경에 능동적이고 신속하게 대처할 수 있게 하는 정보통신기술의 역할과 적용범위 등을 이론강의, 사례연구 및 실습 등을 통하여 학습한다. 비지니스 리엔지니어링(BPR), 전사적 품질경영(TQM), 벤치마킹, 리스트럭춰링, 가상기업, 학습조직, 네트워크조직, 시간기준 경쟁 등의 주제를 다룬다.
   마케팅원론   전사적 마케팅 활동에 실제적으로 응용할 수 있는 전략적 접근방법을 연구한다. 마케팅의 기본개념을 이해하고, 마케팅 환경분석과 전략수립 과정, 그리고 마케팅믹스의 실행과정에 대해서 이해한다.   급속한 환경의 변화, 소비자 욕구의 개성화, 경쟁의 심화로 대변되는 현대의 기업상황에서 기업의 목표를 원활히 수행하기 위해 필요한 마케팅적 사고를 익히고 전사적 마케팅 활동에 실제적으로 응용할 수 있는 전략적 접근방법을 연구한다.
   조직행동론   조직 내에서의 인간행동에 대한 원인을 규명하여 조직의 효율성을 증대시키기 위한 방안에 대해 학습한다.   조직경영의 인간측면과 조직측면에 관련된 여러가지 주제의 학습을 통해서 구체적으로 조직행동이론의 역사적인 전개와 환경요소를 이해하고, 조직행동이론의 여러 가지 주제의 학습을 통해 조직 내 인간 행동의 이해의 틀을 구성하고 예측, 관리할 수 있는 지식을 축적하며, 사례연구를 통해 협동과 조화에 바탕을 두고, 객관적으로 사고하며, 문제를 해결할 수 있는 능력과 태도를 개발하고자 한다.
   광고론   광고이론을 체계적으로 이해시키며 타 마케팅관련 변수들과의 유기적 관계를 연구한다.   현대기업의 경영에서 중요한 비중을 차지하는 광고의 이론적 배경과 실무적 기법을 습득케하여 광고를 체계적으로 이해시키며 타 마케팅관련 변수들과의 유기적 관계를 연구한다.
   경영자료처리   기업업무 수행에 필요한 컴퓨터의 작동원리를 이해시키고 실습을 통하여 기업에서 활용가능성이 높은 사무용 소프트웨어의 사용방법을 익히도록 한다.   PC를 중심으로 컴퓨터의 구조, 정보시스템의 개념 및 활용방법을 이해시키기 위하여 ComputerHardware, Software, 데이터 처리방법, File의구성, Data Base, Data Communications 전반에 대한 기본 지식을 소개한다. 또한 기업경영에서 가장 많이 사용되는 표 계산용 소프트웨어(Spreadsheet), 데이터베이스(Database) 등 주요 응용소프트웨어를 실습을 통하여 습득시킨다
   인사관리   인사관리 기본틀을 이해하는데 중점을 두며,우리나라 인사관리의 추이와 대처 방안을 연구한다.   기업의 인적자원을 관리하는 것은 궁극적으로는 업무를 잘해서 경영전략을 효율적으로 수행하는데에 있다. 인적자원관리의 cycle, 임금체계, 직무분석 그리고 노사관계가 주요 내용을 이룬다. 교과서적인 원리와 우리나라 현실의 새로운 경영환경변화를 가늠하면서 우리 실정에 맞는 직무와 인력관리의 기법과 개념을 구체적으로 다룬다.
   서비스마케팅   산업화가 진전됨에 따라 산업구조도 제조업에서 서비스업 중심으로 전환되고 있다. 서비스 산업의 중요성을 살펴보고, 제조업과 달리 서비스 산업에서 마케팅 업무가 어떤 방식으로 운영되어야 하는지에 대해서 이해한다.   서비스의 특성을 살펴보고, 서비스 마케팅의 기본적 개념을 이해한다. 서비스 산업에서의 시장 환경분석, 그리고 서비스 마케팅 믹스에 대해서 공부한다. 기존의 미케팅 믹스 (4P)에 추가적으로 사람(people), 프로세스(process), 물리적 증거(physical evidence)의 관리방법에 대해서 공부한다.
   재무관리   기업의 경영의사결정의 여러 분야 중 자금과 관련된 부문 즉 자금의 조달과 조달된 자금의 투자를 어떻게 효율적으로 할 것인가에 관한 기초적 지식을 배양함을 목적으로 한다.   기업의 자금조달 및 그 운영을 여하히 효율화하여 기업가치의 극대화를 도모할 것인가를 연구하며 이를 위하여 자본구조이론, 자본예산 및 배당 정책등을 학습한다.
   고급재무회계   중급회계에서 다루지 않았던 재무회계의 특수 분야인 고급수준의 회계이론의 근거를 숙지하고 사업결함, 연결, 환율변동, 파생상품 관련 회계실무처리 능력을 함양한다.   본 교과목은 중급회계에서 다루지 않았던 재무회계의 특수분야를 학습한다. 기업의 사업결합 관련 합병회계와 연결회계를 숙지하며, 해외거래에서 발생한 외화자산부채의 평가 및 파생상품 회계 등으로 구성되어 있다. 본 과목은 기업의 유능한 회계실무자가 되기 위해서 뿐만 아니라 공인회계사(CPA),세무사(CTA), 재무분석사(CFA) 등 회계 및 재무 분야의 전문가 시험을 준비하는데 필요한 과목이다
   원가 및 관리회계   제품의 정확한 원가정보를 생성하는 절차와 방법을 습득하고, 이를 바탕으로 경영자의 내부자원관리에 대한 의사결정과 부서 및 개인의 평가를 위해 회계정보를 구별, 측정, 분석하는 것을 습득하는 것을 목표로 한다.   원가의 기초개념 및 원가계산 제도의 한 학습과정을 통해 습득한 원가정보를 기초로 의사결정 중심의 회계정보 활용을 다룸으로써 원가결정과 경제적 의사결정과의 관련성을 보다 확실히 파악할 수 있도록 한다. 이와 아울러 다양한 의사결정방법을 습득케하여 합리적인 의사결정을 도와주는 학습과정이다.
   경영학원론   경영학이론의 발전과정, 경영학의 체계, 현대기업의 주요 경영이론 및 기법, 기업환경의 변화에 따른 경영자의 역할과 책임 등을 배우도록 한다.   경영학의 기초과목으로 자유시장경제의 기반이 되는 기업제도 및 변화하는 기업환경하에서 운영되는 기업의 기본개념, 및 기업경영의 중요기능 분야에 대한 개괄적인 이해를 도모하는데 목적을 둔다. 경영학 전반에 대한 체계와 기초지식을 바탕으로 현대기업의 경영이론과 경영기법을 터득하고 의사결정 지향적인 경영을 배운다.
   상법   기업법으로서의 상법의 지위를 밝히고, 상법이론에 관한 기초를 익힌다. 상법은 영리를 목적으로 하는 기업을 규율하는 법이므로 일반 사법인 민법의 내용과 상당히 다르기 때문에 이를 먼저 이해해야 한다.   법학을 전공하지 않은 경영학과 학생을 위한 교과목이므로 기업경영과 관련된 법조문을 위주로 강의하고 실무에 어떻게 적용될 수 있는지에 중점을 두고 강의한다. 상법 중 회사편을 중심으로 강의하므로 회사에 대한 기본적인 이해가 필요하다.
   상업정보(컴퓨터)교육론   상업정보(컴퓨터)교과교재연구 및 지도법의 목표는 경영을 위한 기초 학문인 상업정보의 효과적인 교육을 위한 교수법을 배우고 실습하는 것이다.   교육학의 기초 이론들에 대한 이해를 바탕으로 상업정보와 관련한 교과서를 분석하고 이를 효과적으로 전달하기 위한 교수법을 학습하고 이를 실습한다.   
   기업경영과사회   자본주의, 정보화사회와 이익사회로의 변화에 대한 이해를 돕는다. 또한 기업윤리에 대한 대처방안을 습득할 뿐 아니라 경영학적 사고를 습득한다.   일반경영학이 기업내부의 운영에 초점이 맞추어져 있다면 기업경영과 사회는 기업을 둘러싼 사회, 경제, 문화, 기술의 환경변화를 직면하여 거시적인 관점에서 이를 어떻게 대처해 가야하는 지를 학습한다. 자본주의의 속성, 제3의 물결과 권력이동, 기업윤리, 기업의 대 사회정책, 한국사회의 변동의 틀 등을 공부한다.  
   경영전략   경영환경에 적응하고 대처하는 경영전략을 연구한다.   경쟁환경에서 기업이 살아남을 수 있는 전략을 배우고, 타 기업과의 협력 방안을 모색한다. 계량적 기법으로 분석하는 능력도 배양한다.
   금융기관론   자금의 융통과 관련된 기관들에 대한 제도적 측면 및 금융기관을 둘러싼 이론적 측면에 대한 강의를 통하여 자금의 흐름에 대한 이해를 깊게 하고 자금운용에 대한 시야를 넓히는 것을 목표로 한다.   흑자주체(가계)와 적자주체(기업)를 연결해 자금의 융통을 담당하는 금융시장에 대한 실무 및 이론은 물론 은행 및 비은행 금융기관등 금융기관에 대한 실무 및 이론에 대한 소양을 습득케하여 금융과 관련된 기관들에 대한 소양을 풍부하게 한다
   경제학원론Ⅰ   경제학원론의 미시경제학 부분으로서 소비자 형태와 기업형태에 대한 기초 개념과 이론들을 소개한다. 그리고 수요와 공급이 이루는 균형 개념과 시장구조에 따른 시장경제의 특성들을 살핀다   - 소비자 관련 이론 : 수요와 효용, 수요와 탄력성  - 생산자 관련 이론 : 공급과 생산공장 공급과 생산비용  - 시장 균형과 경제적 후생 및 정책  - 시장형태 및 시장성공과 실패   1) 퀴
   증권투자론   개인의 입장에서 투자의 효율성 극대화를 위한 지식의 함양을 목표로 하며 특히 증권투자에 관한 이론 및 실무적 측면에 초점을 맞추어 강의함으로써 투자에 관한 기초적 소양을 함양함을 목표로 한다.   증권투자를 합리적으로 할 수 있도록 증권투자에 관한 이론 및 실무에 관한 지식을 습득케하며 특히 증권시장 실무, 효율적 자본시장이론, 자본자산가격 결정모형, 옵션이론, 선물이론 등에 관한 내용을 연구한다
   중급회계Ⅰ   기업에 대한 정보룰 외부의 이해관계자들에게 생산, 전달하는 재무회계의 이론적 근거를 숙지하고 자산관련 회계실무처리 능력을 함양한다.   본 교과목은 회계정보를 산출하는 재무회계의 이론적 근거를 숙지하고 실무처리 능력을 높이도록 회계기준에 의한 회계처리 지침의 내용과 이론적 근거 및 회계처리 실무 내용을 학습한다. 본 과목을 수강하기 위해서는 회계원리에 대한 선행학습이 요구되며, 교과 내용은 중급회계의 전반부인 재무회계의 이론적 구조 및 자산회계의 이론적 회계처리 등으로 구성되어 있다. 본 과목은 경영자로서 기본적으로 알아야 할 재무회계에 대한 개념 및 기본지식을 전 달할 뿐만 아니라 공인회계사(CPA), 세무사(CTA),재무분석사(CFA)등 회계 및 재무 분야의 전문가 시험을 준비하는데 필요한 과목이다
   마케팅조사론   주어진 기업환경하에서 마케팅 전략수립에 필요한 시장조사를 직접 수행할 수 있는 능력을 도모한다. 마케팅 의사결정에 도움을 줄 수 있는 정보를 수집하고, 분석, 제시할 수 있는 방법론을 공부한다.   합리적 마케팅 의사결정에 유용한 정보를 줄 수 있는 시장조사를 행하기 위해 알아야 할 시장조사의 절차 및 수행과정, 실험설계, 분석적 통계기법들을 연구하여, 시장분석과 예측, 소비자 조사, 판매예측 등 동태적 마케팅활동과 관련되는 제 문제에 체계적으로 접근한다.(선수과목 : 통계학)
  경공A000  회계원리   경영활동 전반에 대한 회계정보를 생성하기 위해 서 세계 공통의 기업언어인 회계처리방법을 익혀야 한다. 이를 위하여 거래나 사건을 인식, 측정, 분류하여 이를 회계정보로 전달하는 방법을 습득하여야 한다. 본 과목은 회계의 개념, 회계원칙과 회계목적에 대한 이해와 함께 재무제표의 작성과 이해에 대한 기초적인 지식을 배양하는 데 있다.   회계원리는 회계학의 기초과목으로서 거래의 인식과 측정, 분류 및 재무제표 작성 방법을 학습한다. 회계원리는 회계과목의 출발점으로 기업과 관련된 사건과 거래를 회계처리하고 표현하는 방법을 배우는 과정이다.
   경영수리모델분석   수학과 통계학 등 수리적인 기법을 활용하여 경영분야에 응용하는 기법을 배운다.   수학과 통계학을 기초로 한 회귀분석이나 t-test 등의 기법을 우선 공부한다. 이를 활용하여 경영분야에 적용하는 기법과 사례를 연구한다.
   전략적 브랜드 관리   현대사회에서의 브랜드의 중요성을 인식하고, 브랜드의 가치를 증가시키는 브랜드 관리에 대한 전략적 사고를 심어주고자 한다. 또한 브랜드가 사회와 기업에 미치는 영향에 대해서도 함께 살펴보고자 한다.   브랜드의 중요성에 대해 이해하고, 브랜드관리에 관련된 이론적 내용으로 브랜드 컨셉 개발, 브랜드 아이덴티티 전략, 브랜드 이미지 관리, 그리고 브랜드 자산의 구축, 활용, 강화등에 대해 다룬다. 그 외에 현실적인 사례등을 통해 현실에서 적용되고 있는 전략들을 검토하고, 직접적으로 브랜드 전략을 구축해보는 프로제트도 수행해본다.(선수과목 : 광고론
   경영분석   기업과 관련된 이해관계자들 즉 투자자, 금융기관, 세무당국, 경영자 등이 특정 기업에 관한 그들의 경제적 의사결정을 합리적이고 과학적인 기초에서 할 수 있는 능력을 배양함을 목적으로 한다.   기업의 양적, 질적 요인에 대한 분석을 통하여 기업의 재무상태 및 경영성과를 과학적으로 판단할 수 있는 능력을 함양토록하여 증권투자등 기업과 관련된 의사결정을 합리적으로 할 수 있는 소양을 갖추도록 한다.
   경영분석   기업과 관련된 이해관계자들 즉 투자자, 금융기관, 세무당국, 경영자 등이 특정 기업에 관한 그들의 경제적 의사결정을 합리적이고 과학적인 기초에서 할 수 있는 능력을 배양함을 목적으로 한다.   기업의 양적, 질적 요인에 대한 분석을 통하여 기업의 재무상태 및 경영성과를 과학적으로 판단할 수 있는 능력을 함양토록하여 증권투자등 기업과 관련된 의사결정을 합리적으로 할 수 있는 소양을 갖추도록 한다.
  e-비즈니스  e-비즈니스   디지털 경제의 핵심이 되고 있는 e-비즈니스와 전자상거래 응용에 대한 이론과 실제를 학습하고 경영자로서 관련 이슈에 대한 전략적 대응 능력을 배양하는데 그 목표를 둔다.   인터넷을 비롯한 통신 네트워크를 통하여 일어나는 기업 경영상의 활동과 그 의의를 학습하며, e-비즈니스 및 전자적 상거래가 기업에 전략적으로 미치는 영향과 활용형태를 검토하고 이를 통하여 기업목표를 효과적으로 달성하기 위한 방안을 살펴본다. 본 강좌에서는 전자상거래, m-코머스, u-코머스와 관련된 기본 기술과 이론에 대하여 강의를 한다.
   파생상품론   파생상품중 대표적인 상품인 옵션과 선물에 중점을 두어 학습하며 파생상품에 대한 이론과 실제투자에 대한 지식을 동시에 함양한다.   본 과목은 재무관리분야에 있어 최근에 특히 그 중요성이 강조되고 있는 파생상품에 대해 학습한다. 특히 주가지수옵션을 비롯한 옵션 전반에 대한 이론적 탐구와 동시에 실제 옵션투자에 대한 능력을 증진시킨다. 또한 주가지수 선물을 비룻한 선물전반에 대한 이론적 연구와 동시에 실제 선물투자에 대한 능력을 함양한다.  본 과목은 선물거래상담사, 재무위험관리사(FRM), 금융자산관리사(FP)등의 자격시험을 준비하는데도 유용한 과목이다
   전략경영모델   한국 기업에 대한 종합적인 이해력을 증진할 뿐 아니라 창의적인 사고를 개발한다. 또한 이슈가 되고 있는 경영학 문제를 습득하여 취직에 대비한다.   신경영기법으로 불리우는 현재 우리나라 기업에서 시행되는 신인사제도, 기업문화, TQM, BPR, TQC, ABC, 팀제, 린생산방식 등을 한국사회변동을 가늠하면서 종합적인 모델로 유기적으로 체계화하여 실천적인 대안을 마련한다. 기업과 사회, 인사관리, 조직행동에서 부분적으로 언급된 내용과 경영학 전반에서 현재 우리나라 기업들이 현안으로 다루고 있는 사안들을 실무경험과 조사를 바탕으로 핵심적으로 통합하였다. 취직시험과 면담을 앞두고 경영학 전반을 점검한다. 
"  
# <- doc1 은 전공소개 전체를 넣었는데 학과소개로 바꿀까 생각중 

doc2 = "ncs_cap의 각 행" 
```
